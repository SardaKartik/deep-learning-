for high accuracy in machine learning our values are scale

if we divide the whole array by 255 it will be scaled in between 0 to 1 since the value lies in between 0 to 255

```python
x_train=x_train/255
```

```python
x_test=x_test/255
```

```python
x_train_flattened =x_train.reshape(len(x_train),28*28)
```

in reshape we want our 60000 data and secod dimensional that is 28*28

```python
x_test_flattened=x_test.reshape(len(x_test),28*28)
```

```python
x_train_flattened[0]
```

creating a simple neural network 

it has just two layer input layer with 784 elements output layer 10 element 

```python
model=keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),
                       activation='sigmoid')
])
```

sequential meaning  i am having a stack of layer in a neural network 
since it is stack every layer as one element 

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

sparse_categorical_crossentropy means is our output class are categorical because we have categorical bassically 0 to 9 we have like 10 classes in our output and sparse means our output variable which is y train is actually an integer int number if it is one hot encoded array you use probably use categorical cross entrophy 

```python
model.fit(x_train_flattened,y_train,epochs=11)
```

here we notice that as the time goes the accuracy is increasing 

here it is evaluating the accuracy on the trained data set 

```python
model.evaluate(x_test_flattened,y_test)
```

evaluating accuracy on the test data set how much accurate our model is 

```python
model.predict(x_test_flattened)
```

now we will import confusion matrix from tensorflow we call do use sklearn library 

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_label)
```

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel("predicted")
plt.ylabel("truth")
```

annotbool or rectangular dataset, optional

now we will add hidden layer in our model 
and then show how it affect on our output 

generally hidden layer improve the performance 

```python
model = keras.Sequential([
    #keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(100, activation='relu'),# specificy no of neuron in hidden layer 
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train_flattened,y_train,epochs=11)
```

one epoch mean that we have go through all the training sample once  and cacluste the loss 
and here we hae use the 11 epoch whcih mean we have been gone through our training set 11 time 

```python
model.evaluate(x_test_flattened,y_test)
```

```python
y_predicted=model.predict(x_test_flattened)
```

```python
y_predicted_label=[np.argmax(i) for i in y_predicted]
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_label)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel("predicted")
plt.ylabel("truth")
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)), # we use this function to flatten our array 
    keras.layers.Dense(100, activation='relu'),# specificy no of neuron in hidden layer 
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train,y_train,epochs=11) # as we have flattened our array 
```
