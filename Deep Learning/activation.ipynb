```python
import math 
```

```python
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(1)
```

```python
sigmoid(0)
```

```python
sigmoid(-4.25)
```

```python
# tanh 
```

```python
def tanh(x):
    return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))
```

```python
tanh(0)
```

```python
tanh(15)
```

```python
tanh(-15)
```

```python
def relu(x):
    return max(0,x)
```

```python
relu(5)
```

```python
relu(3)
```

```python
relu(-2)
```

```python
# leaky relu
```

```python
def relu(x):
    return max(0.1*x,x)
```

```python
relu(5)
```

```python
relu(0.12)
```

```python
relu(-2.3)
```
